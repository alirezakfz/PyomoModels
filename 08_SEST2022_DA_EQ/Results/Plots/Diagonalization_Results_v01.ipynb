```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path='Diagonalization_Data/diagonalizaton_results_'
df_fixed_evs_ahe = pd.read_csv(file_path+'Fixed_EVs_AHE.csv')
df_fixed_pvs_ahe = pd.read_csv(file_path+'Fixed_PVs_AHE.csv')

df_fixed_evs_diag = pd.read_csv(file_path+'Fixed_EVs_DIAG.csv')
df_fixed_pvs_diag = pd.read_csv(file_path+'Fixed_PVs_DIAG.csv')

df_mixed_ahe=  pd.read_csv(file_path+'mixed_AHE.csv')
df_mixed_diag =  pd.read_csv(file_path+'mixed_DIAG.csv')
```

```python
df_all = [df_fixed_evs_ahe, df_fixed_pvs_ahe, df_fixed_evs_diag, df_fixed_pvs_diag, df_mixed_ahe, df_mixed_diag]
```

```python
for df in df_all:
    df.drop(df[df.Iteration==0].index, inplace=True)
    df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df_fixed_evs_ahe.groupby('Time').mean().drop(columns=['Iteration']).reset_index()
#df_fixed_evs_ahe
```

```python
offer_col = ['offer_0'+str(x) for x in range(1,10)]
time_offer =['Time']
time_offer = [*offer_col, *time_offer]
```

```python
df_fixed_evs_ahe.groupby('Time').mean().drop(columns=['Iteration']).reset_index()[time_offer].head()
```

```python
df_fixed_evs_ahe.groupby('Time').mean().drop(columns=['Iteration']).reset_index()[offer_col].head()
```

```python
df_melt = pd.melt(df_fixed_evs_ahe.groupby('Time').mean().drop(columns=['Iteration']).reset_index()[time_offer], 
          id_vars=['Time'], 
          value_vars=offer_col )
```

```python
df_melt.head()
```

```python
sns.histplot(data=df_melt, stat='density',  multiple="stack",
             x="Time", kde=False,
             palette="pastel", hue="variable",
             element="bars", legend=True) #stat="value",
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(1,1,1)
sns.barplot(data=df_melt, x="Time",y='value',hue="variable", ax=ax )
```
