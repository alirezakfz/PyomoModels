```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path='Diagonalization_Data/diagonalizaton_results_'

df_fixed_evs_diag = pd.read_csv(file_path+'Fixed_EVs_DIAG.csv')
df_fixed_pvs_diag = pd.read_csv(file_path+'Fixed_PVs_DIAG.csv')
df_mixed_diag =  pd.read_csv(file_path+'mixed_DIAG.csv')
```

```python
df_all = [ df_mixed_diag, df_fixed_evs_diag, df_fixed_pvs_diag]
data_order=['Mixed','Fixed_EVs', 'Fixed_PVs']
```

```python
for df in df_all:
    df.drop(df[df.Iteration==0].index, inplace=True)
    df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df_fixed_evs_diag.groupby('Time').mean().drop(columns=['Iteration']).reset_index()
```

```python
offer_col = ['offer_0'+str(x) for x in range(1,10)]
time_offer =['Time']
time_offer = [*offer_col, *time_offer]
```

```python
df_fixed_evs_diag.groupby('Time').mean().drop(columns=['Iteration']).reset_index()[time_offer].head()
```

```python
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
```

# Offers Bids Plots

### Offers Bids Plots: All Scenarios All DAs based on time interval

```python
counter = 0
offer = []
scenario=[]
time=[]
#iterate trough all dataframes and average their offers bids : all DAs in each 3 scenario
for df in df_all:
    temp = df.groupby('Time').mean().drop(columns=['Iteration']).reset_index()[time_offer].drop('Time',axis=1).mean(axis=1)
    count=16
    for i in temp:
        offer.append(i)
        scenario.append(data_order[counter])
        time.append(count)
        count+=1
    counter+=1

df_compare = pd.DataFrame({'Offers':offer, 'Scenario':scenario, 'Time':time})
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(1,1,1)

sns.barplot(data= df_compare, x='Time',y='Offers' ,hue='Scenario', ax=ax);

ax.set_title('Average Offers Bids In ALL Scenarios for all DAs', fontsize=12)
ax.set_xlabel('Time Interval',fontsize=12);
ax.set_ylabel('Offers Bid in Megaatt PU',fontsize=12);

ax.plot()

ax.set_rasterized(True)
fig.savefig("Figures_Diganolization/Average_Offers_ALL_Scenarios.png", bbox_inches= 'tight', dpi=200)
```

### Offers Bids Plots: All Scenarios All DAs in horizon

```python

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)

sns.barplot(data= df_compare.groupby("Scenario").mean().reset_index().drop('Time',axis=1),
            x='Scenario',y='Offers' , ax=ax);

ax.set_title('Average Offers Bids In ALL Scenarios for all DAs', fontsize=14)
ax.set_xlabel('Scenarios',fontsize=14);
ax.set_ylabel('Offers Bid in Megaatt PU',fontsize=14);

change_width(ax, .45)

ax.set_yticklabels(ax.get_yticks().round(2), size = 12)
ax.set_xticklabels(ax.get_xticklabels(), size = 12)

ax.plot()

ax.set_rasterized(True)
fig.savefig("Figures_Diganolization/Average_Offers_ALL_Scenarios_Horizon.png", bbox_inches= 'tight', dpi=200)
```

### Offers Bids Plots: All Scenarios Based ON DAs Group
There is 3 groups for DAs having different EVs and PVs penetration
- group_01 = 2,4,6,8
- group_02 = 3,9
- group_03 = 1,5,7

```python
offer_col = ['offer_0'+str(x) for x in range(1,10)]
time_offer =['Time']
time_offer = [*offer_col, *time_offer]


DA = dict()
for i in range(1,10):
    if i % 2 == 0:
        DA['offer_0'+str(i)] = 'CAT_01'
    elif i % 3 == 0:
        DA['offer_0'+str(i)] = 'CAT_02'
    else:
        DA['offer_0'+str(i)] = 'CAT_03'

#offer = 
# scenario=[] There is one above
#time=[16+i for i in range(24)]


first_iteration = True

for df in df_all:
    temp= df.groupby('Time').mean().drop(columns=['Iteration']).reset_index()[time_offer]
    if first_iteration:
        df_compare = temp.copy(deep=True)
        first_iteration=False
    else:
        df_compare = pd.concat([df_compare, temp])

# Take the average based on time
df_compare = df_compare.groupby('Time').mean().reset_index()

df_compare = df_compare.melt(id_vars=['Time'], var_name="Category", value_name="Offers")

for key, value in DA.items():
    df_compare.loc[df_compare['Category'] == key, 'Category'] = value

df_compare = df_compare.groupby(['Time','Category']).mean().reset_index()
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(1,1,1)

sns.barplot(data= df_compare, x='Time',y='Offers' ,hue='Category', ax=ax);

ax.set_title('Average Offers Bids In ALL Scenarios Based on DAs Category', fontsize=12)
ax.set_xlabel('Time Interval',fontsize=12);
ax.set_ylabel('Offers Bid in Megaatt PU',fontsize=12);


ax.plot()

ax.set_rasterized(True)
fig.savefig("Figures_Diganolization/Average_Offers_DAs_Categories.png", bbox_inches= 'tight', dpi=200)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)

sns.barplot(data= df_compare.groupby("Category").mean().reset_index().drop('Time',axis=1),
            x='Category',y='Offers' , ax=ax);

ax.set_title('Average Offers Bids In ALL Scenarios for all Categories', fontsize=14)
ax.set_xlabel('Time Interval',fontsize=14)
ax.set_ylabel('Offers Bid in Megaatt PU',fontsize=14)

change_width(ax, .45)

ax.set_yticklabels(ax.get_yticks().round(2), size = 14)
ax.set_xticklabels(ax.get_xticklabels(), size = 14)

ax.plot()

ax.set_rasterized(True)
fig.savefig("Figures_Diganolization/Average_Offers_ALL_Categories_Horizon.png", bbox_inches= 'tight', dpi=200)
```
