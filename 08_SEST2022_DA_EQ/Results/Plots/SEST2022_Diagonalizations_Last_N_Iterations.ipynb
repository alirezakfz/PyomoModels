```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path='Diagonalization_Data/diagonalizaton_results_'

df_fixed_evs_diag = pd.read_csv(file_path+'Fixed_EVs_DIAG.csv')
df_fixed_pvs_diag = pd.read_csv(file_path+'Fixed_PVs_DIAG.csv')
df_mixed_diag =  pd.read_csv(file_path+'mixed_DIAG.csv')
```

```python
df_all = [ df_mixed_diag, df_fixed_evs_diag, df_fixed_pvs_diag]
data_order=['Mixed','Fixed_EVs', 'Fixed_PVs']
```

```python
for df in df_all:
    df.drop(df[df.Iteration==0].index, inplace=True)
    df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df_fixed_evs_diag.groupby('Time').mean().drop(columns=['Iteration']).reset_index()
```

```python
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
```

# Offers Bids Plots

```python
demand_col = ['bids_0'+str(x) for x in range(1,10)]
time_offer =['Time']
time_offer = [*demand_col, *time_offer]

counter = 0
demand = []
scenario=[]
time=[]

# how many last iterations to show
last_iterations = 50
no_iterations = 300

#iterate trough all dataframes and average their offers bids : all DAs in each 3 scenario
for df in df_all:
    df = df[df['Iteration'] > no_iterations - last_iterations]
    temp = df.groupby('Time').mean().drop(columns=['Iteration']).reset_index()[time_offer].drop('Time',axis=1).mean(axis=1)
    count=16
    for i in temp:
        demand.append(i)
        scenario.append(data_order[counter])
        time.append(count)
        count+=1
    counter+=1

df_compare = pd.DataFrame({'Demands':demand, 'Scenario':scenario, 'Time':time})


#**********************************************************
offer_col = ['offer_0'+str(x) for x in range(1,10)]
time_offer =['Time']
time_offer = [*offer_col, *time_offer]

counter = 0
offer = []
scenario=[]
time=[]
#iterate trough all dataframes and average their offers bids : all DAs in each 3 scenario
for df in df_all:
    temp = df.groupby('Time').mean().drop(columns=['Iteration']).reset_index()[time_offer].drop('Time',axis=1).mean(axis=1)
    count=16
    for i in temp:
        offer.append(i)
        scenario.append(data_order[counter])
        time.append(count)
        count+=1
    counter+=1

df_offer = pd.DataFrame({'Offers':offer, 'Scenario':scenario, 'Time':time})

df_compare['Offers'] = df_offer['Offers']

df_compare = df_compare.melt(id_vars=['Time','Scenario'], var_name="Bid_Type", value_name="Bid_Value")
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)

sns.barplot(data= df_compare.groupby(["Scenario","Bid_Type"]).mean().reset_index().drop('Time',axis=1),
            x='Scenario',y='Bid_Value', hue='Bid_Type' , ax=ax)

ax.set_title('Average Demands and Offers Bids In Last %d Iterations' %(last_iterations), fontsize=14)
ax.set_xlabel('Scenarios',fontsize=14);
ax.set_ylabel('Average Bids in Megaatt PU',fontsize=14);

# change_width(ax, .45) 

# ax.set_yticklabels(ax.get_yticks().round(2), size = 12)
ax.set_xticklabels(ax.get_xticklabels(), size = 12)

ax.plot()

ax.set_rasterized(True)
fig.savefig("Figures_Diganolization/Average_Demands_and_Offer_Bids_Scenarios_Horizon_last_n_Iterations.eps", bbox_inches= 'tight', dpi=200)
```

# Based on Category and Scenarios

```python
DA_demand = dict()
for i in range(1,10):
    if i % 2 == 0:
        DA_demand['bids_0'+str(i)] = 'CAT_01'
    elif i % 3 == 0:
        DA_demand['bids_0'+str(i)] = 'CAT_02'
    else:
        DA_demand['bids_0'+str(i)] = 'CAT_03'
        
DA_offer = dict()
for i in range(1,10):
    if i % 2 == 0:
        DA_offer['offer_0'+str(i)] = 'CAT_01'
    elif i % 3 == 0:
        DA_offer['offer_0'+str(i)] = 'CAT_02'
    else:
        DA_offer['offer_0'+str(i)] = 'CAT_03'

        
# how many last iterations to show
last_iterations = 50
no_iterations = 300


#**************************************************************
first_iteration = True
counter = 0

for df in df_all:
    df = df[df['Iteration'] > no_iterations - last_iterations]
    temp= df.groupby('Time').mean().drop(columns=['Iteration']).reset_index()
    temp['Scenario']= data_order[counter]
    if first_iteration:
        df_compare = temp.copy(deep=True)
        first_iteration=False
        
    else:
        df_compare = pd.concat([df_compare, temp], axis=0)
    counter += 1    

df_compare = df_compare.reset_index()
df_compare.drop(['index', 'Time'],axis=1, inplace=True)


df_compare = df_compare.groupby('Scenario').mean().reset_index()


#**************************************************************************
offer_col = ['offer_0'+str(x) for x in range(1,10)]
time_offer =['Scenario']
time_offer = [*offer_col, *time_offer]

df_offer = df_compare[time_offer]

#**********************************************************************
demand_col = ['bids_0'+str(x) for x in range(1,10)]
time_demand =['Scenario']
time_demand = [*demand_col, *time_demand]

df_demand = df_compare[time_demand]

#****************************************************************
df_offer = df_offer.melt(id_vars=['Scenario'], var_name="DA_Category", value_name="Bid_Value")

df_demand = df_demand.melt(id_vars=['Scenario'], var_name="DA_Category", value_name="Bid_Value")

#***************************************************************
for key, value in DA_demand.items():
    df_demand.loc[df_demand['DA_Category'] == key, 'DA_Category'] = value

df_demand = df_demand.groupby(['Scenario','DA_Category']).mean().reset_index()

for key,value in DA_offer.items():
    df_offer.loc[df_offer['DA_Category'] == key, 'DA_Category'] = value

df_offer = df_offer.groupby(['Scenario','DA_Category']).mean().reset_index()
#****************************************************************

# fig = plt.figure(figsize=(8,6))
# # axs = fig.add_subplot(1,2)
# axs = plt.subplots(1, 2)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.barplot(data= df_demand,
            x='DA_Category', y='Bid_Value', hue='Scenario' , ax=axes[0])
axes[0].set_title("Average Demand Bids Last %d Iterations"%(last_iterations), fontsize=12)
axes[0].set_xlabel('DAs Categories',fontsize=14);
axes[0].set_ylabel('Average Demands Bids MW-PU',fontsize=14);


sns.barplot(data= df_offer,
            x='DA_Category', y='Bid_Value', hue='Scenario' , ax=axes[1])

axes[1].set_title("Average Offers Bids Last %d Itarations"%(last_iterations), fontsize=12)
axes[1].set_xlabel('DAs Categories',fontsize=14);
axes[1].set_ylabel('Average Offers Bids MW-PU',fontsize=14);

fig.suptitle('Average Demands and Offers Bids In ALL Scenarios based on DAs Category',fontsize=14)



#axes.set_title(, )
# ax.set_xlabel('Scenarios',fontsize=14);
# ax.set_ylabel('Average Bids in Megaatt PU',fontsize=14);

# # change_width(ax, .45) 

# # ax.set_yticklabels(ax.get_yticks().round(2), size = 12)
# ax.set_xticklabels(ax.get_xticklabels(), size = 12)

plt.show()
fig.set_rasterized(True)
fig.savefig("Figures_Diganolization/Average_Demands_and_Offer_Bids_Scenarios_DAs_Category_Last_n_Iterations.eps", bbox_inches= 'tight', dpi=200)


# # set labels
# plt.setp(axs[-1, :], xlabel='x axis label')
# plt.setp(axs[:, 0], ylabel='y axis label')
```
