```python
%matplotlib inline

import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_diag = pd.read_csv("diagonalizaton_results_diag.csv")
df_ahe = pd.read_csv("diagonalizaton_results_AHE.csv")
df_fp = pd.read_csv("diagonalizaton_results_FP.csv")
df_sfp = pd.read_csv("diagonalizaton_results_SFP.csv")
```

## DA1 in algorithms

```python
bid_col = "bids_01"
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for t in range(16,40):
    bids_t = df_ahe[df_ahe.Time==t][bid_col][291:]
    plt.plot(range((len(bids_t))), bids_t)
    


# plt.plot(range(len(pros1)), pros1)
# plt.plot(range(len(pros2)), pros2)
ax.set_ylabel("Bids for each 24 of time slots in MWh")
ax.set_xlabel("Epochs number")
ax.set_title('The last 10 epochs for DA 1 in AH algorithm')
# ax.legend(['prosumer 1', 'prosumer 2'])
ax.set_xticklabels(range(291,292+len(bids_t)))
plt.savefig("DA1_10_last_epochs_bids_AH.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for t in range(16,40):
    bids_t = df_fp[df_fp.Time==t][bid_col][145:]
    plt.plot(range((len(bids_t))), bids_t)
    


# plt.plot(range(len(pros1)), pros1)
# plt.plot(range(len(pros2)), pros2)
ax.set_ylabel("Bids for each 24 of time slots in MWh")
ax.set_xlabel("Epochs number")
ax.set_title('The last 10 epochs for DA 1 in FP algorithm')
ax.set_xticklabels(range(145,146+len(bids_t)))
# ax.legend(['prosumer 1', 'prosumer 2'])
plt.savefig("DA1_10_last_epochs_bids_FP.eps")
plt.savefig("DA1_10_last_epochs_bids_FP.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for t in range(16,40):
    bids_t = df_sfp[df_sfp.Time==t][bid_col][291:]
    plt.plot(range((len(bids_t))), bids_t)
    


# plt.plot(range(len(pros1)), pros1)
# plt.plot(range(len(pros2)), pros2)
ax.set_ylabel("Bids for each 24 of time slots in MWh")
ax.set_xlabel("Epochs number")
ax.set_title('The last 10 epochs for DA 1 in SFP algorithm')
ax.set_xticklabels(range(291,292+len(bids_t)));
# ax.legend(['prosumer 1', 'prosumer 2'])
# plt.savefig("DA1_10_last_epochs_bids_FP.eps")
plt.savefig("DA1_10_last_epochs_bids_SFP.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for t in range(16,40):
    bids_t = df_diag[df_diag.Time==t][bid_col]
    plt.plot(range((len(bids_t))), bids_t)
    


# plt.plot(range(len(pros1)), pros1)
# plt.plot(range(len(pros2)), pros2)
ax.set_ylabel("Bids for each 24 of time slots in MWh")
ax.set_xlabel("Epochs number")
ax.set_title('The last 6 epochs for DA 1 in SFP')
# ax.set_xticklabels(range(145,146+len(bids_t)))
# ax.legend(['prosumer 1', 'prosumer 2'])
# plt.savefig("DA1_10_last_epochs_bids_diag.eps")
# plt.savefig("DA1_10_last_epochs_bids_diag.png")
```
