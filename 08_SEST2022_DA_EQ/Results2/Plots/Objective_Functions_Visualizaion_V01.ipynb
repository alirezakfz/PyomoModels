```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1) Loading Data
There are 6 files belonging to 3 scenarios for two algorithms.

```python
file_path = "Objective_Functions/objective_"


# Diagonalization algorithm
df_diag_mixed = pd.read_csv(file_path + "Mixed_Diag.csv")
df_diag_PVs = pd.read_csv(file_path + "Fixed_PVs_Diag.csv")
df_diag_EVs = pd.read_csv(file_path + "Fixed_EVs_Diag.csv")

# AHE algorithm
df_ahe_mixed = pd.read_csv(file_path + "Mixed_AHE.csv")
df_ahe_PVs = pd.read_csv(file_path + "Fixed_PVs_AHE.csv")
df_ahe_EVs = pd.read_csv(file_path + "Fixed_EVs_AHE.csv")
```

```python
# add all dataframes 
df_all = [df_diag_mixed, df_diag_PVs, df_diag_EVs, df_ahe_mixed, df_ahe_PVs, df_ahe_EVs]

# Drop first iteration values as it run by random values
for df in df_all:
    df.drop(df.index[0],inplace=True)
```

# 2) Compare Dataframe values

### 2.1) Compare based on scenario and algorithm last Iteration

```python
# Compare each last row of data frames with each other as bar plot
cost=[]
da=[]
algo=[]

for idx , val in zip(df_diag_mixed.iloc[-1].index, df_diag_mixed.iloc[-1]):
    cost.append(val)
    da.append('DA'+str(idx))
    algo.append('Diag')

for idx , val in zip(df_ahe_mixed.iloc[-1].index, df_ahe_mixed.iloc[-1]):
    cost.append(val)
    da.append('DA'+str(idx))
    algo.append('AHE')

df_temp = pd.DataFrame({'Cost':cost, 'Algorithm':algo, 'DA':da})
df_temp.Cost = pd.to_numeric(df_temp.Cost,  errors='coerce')

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)

sns.barplot(data= df_temp, x='DA',y='Cost' ,hue='Algorithm', ax=ax);

ax.set_title('Last iteration value of the Objective function for each DA\n MIXED Scenario', fontsize=12)
ax.set_xlabel('Participant Demand Aggregators (DA) in EPEC',fontsize=12);
ax.set_ylabel('Result of Objective functions as COST',fontsize=12);

ax.plot()

ax.set_rasterized(True)
fig.savefig("Figures/obj_figures/Mixed_Last_Iterations_BarPlot_Diag_AHE.eps", bbox_inches= 'tight')
```

### 3.1) Checking status of the diagonalization algorithm

```python
def df_line_plot(df, iter_len, scenario, algorithm):
    fig = plt.figure(figsize=(8,5))
    ax = fig.add_subplot(1,1,1)

    tick_x = [x for x in range(len(df)+iter_len, len(df)+1)]
    
    # Get list of columns and sort them
    list_col =df.columns.tolist()
    list_col.sort()
    list_col

    df.iloc[iter_len:].plot(ax=ax)
    ax.legend(bbox_to_anchor=(1.01, 1),labels=['DA'+cl for cl in list_col]);
    ax.set_title(scenario + ' '+ algorithm + ' '+ 'Algorithm', fontsize=13)
    ax.set_xlabel('Last '+ str(-iter_len)+ ' Iteration of '+algorithm +' '+'Algorithm', fontsize=12)
    ax.set_ylabel('Net Cost', fontsize=12)
    ax.set_xticks(tick_x)
    ax.plot()
    plt.show()
    return ax, fig
```

```python
algorithm = 'Diagonalization'
scenario = 'Mixed'
iter_len = -20
# df_all = [df_diag_mixed, df_diag_PVs, df_diag_EVs, df_ahe_mixed,df_ahe_PVs,df_ahe_EVs]

ax, fig = df_line_plot(df_diag_mixed, iter_len, scenario, algorithm)

ax.set_rasterized(True)
fig.savefig("Figures/obj_figures/Mixed_diag_line_plot.eps", bbox_inches= 'tight')
```
