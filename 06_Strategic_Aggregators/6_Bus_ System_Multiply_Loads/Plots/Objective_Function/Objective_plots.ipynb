```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df_ahe = pd.read_csv("objective_AHE.csv")
df_diag = pd.read_csv("objective_diag.csv")
df_fp = pd.read_csv("objective_FP.csv")
df_sfp = pd.read_csv("objective_SFP.csv")
```

## AHE algorithm plot

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
ax = fig.add_subplot(111)
da_legend=[]

for da in range(1,len(df_ahe.columns)+1):
    plt.plot(range(len(df_ahe)) , df_ahe[str(da)], label="DA_"+str(da))
    da_legend.append(da)

# ax.legend(bbox_to_anchor=(1.04,1), loc='upper center')


# ax.legend(loc='lower center', bbox_to_anchor=(0.5, 1.05),
#           ncol=4, fancybox=True, shadow=True)

ax.legend(loc='upper center', bbox_to_anchor=(0.5, 0.2),
          fancybox=True, shadow=True, ncol=5)
```

## FP algorithm plot

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
ax = fig.add_subplot(111)
da_legend=[]

for da in range(1,len(df_fp.columns)+1):
    plt.plot(range(len(df_fp)) , df_fp[str(da)], label="DA_"+str(da))
    da_legend.append(da)

ax.legend(bbox_to_anchor=(1.08,1), loc='upper center')


# ax.legend(loc='lower center', bbox_to_anchor=(0.5, 1.05),
#           ncol=4, fancybox=True, shadow=True)

# ax.legend(loc='upper center', bbox_to_anchor=(0.5, 0.2),
#           fancybox=True, shadow=True, ncol=5)
```

## SFP Algorithm plot

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
ax = fig.add_subplot(111)
da_legend=[]

for da in range(1,len(df_sfp.columns)+1):
    plt.plot(range(len(df_sfp)) , df_sfp[str(da)], label="DA_"+str(da))
    da_legend.append(da)

ax.legend(bbox_to_anchor=(1.08,1), loc='upper center')


# ax.legend(loc='lower center', bbox_to_anchor=(0.5, 1.05),
#           ncol=4, fancybox=True, shadow=True)

# ax.legend(loc='upper center', bbox_to_anchor=(0.5, 0.2),
#           fancybox=True, shadow=True, ncol=5)
```
