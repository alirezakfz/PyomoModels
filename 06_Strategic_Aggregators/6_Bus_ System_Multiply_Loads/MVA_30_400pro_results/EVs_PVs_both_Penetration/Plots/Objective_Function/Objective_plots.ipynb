# Effect of EVs and Solar on Net Cost 

- DAs 1,5,7 have 20% solar and 35% EVS
- DAs 2,4,6,8 have 50% solar and 75% EVS
- DAs 3,9 have 30% solar and 50% EVS

```python
df_ahe = pd.read_csv("objective_AHE.csv")
df_diag = pd.read_csv("objective_diag.csv")
df_fp = pd.read_csv("objective_FP.csv")
# df_sfp = pd.read_csv("objective_SFP.csv")
```

```python
df_ahe.columns
```

```python
plt.bar(['DA'+ str(x) for x in df_ahe.columns], df_ahe.mean())
```

```python
DAs_list = []
Algo_list =[]
Penetratio_s=[]
Penetration_ev=[]
Average =[]

for col in df_ahe.columns:
    DAs_list.append('DA'+col)
    
    if int(col)%2 == 0:
        Penetratio_s.append('50%')
        Penetration_ev.append('75%')
    elif int(col)%3 ==0:
        Penetratio_s.append('30%')
        Penetration_ev.append('50%')
    else:
        Penetratio_s.append('20%')
        Penetration_ev.append('35%')
    
    Algo_list.append('AHE')

for m in df_ahe.mean():
    Average.append(m)
        
```

```python
for col in df_fp.columns:
    DAs_list.append('DA'+col)
    
    if int(col)%2 == 0:
        Penetratio_s.append('50%')
        Penetration_ev.append('75%')
    elif int(col)%3 ==0:
        Penetratio_s.append('30%')
        Penetration_ev.append('50%')
    else:
        Penetratio_s.append('20%')
        Penetration_ev.append('35%')
    
    Algo_list.append('FP')

for m in df_fp.mean():
    Average.append(m)
```

```python
for col in df_sfp.columns:
    DAs_list.append('DA'+col)
    
    if int(col)%2 == 0:
        Penetratio_s.append('50%')
        Penetration_ev.append('75%')
    elif int(col)%3 ==0:
        Penetratio_s.append('30%')
        Penetration_ev.append('50%')
    else:
        Penetratio_s.append('20%')
        Penetration_ev.append('35%')
    
    Algo_list.append('SFP')

for m in df_sfp.mean():
    Average.append(m)
```

```python
dic_temp = {'Average': Average,
           'Solar_Penetration':Penetratio_s,
           'EVs_Penetration' : Penetration_ev,
           'Algorithm' : Algo_list,
           'DAs' : DAs_list}
```

```python
df_mean = pd.DataFrame.from_dict(dic_temp)
```

```python
df_mean = df_mean.groupby(['Solar_Penetration','EVs_Penetration','Algorithm']).mean().round(2).reset_index()
```
